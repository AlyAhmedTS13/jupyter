```python
df.set_index('ID',inplace=True)
```

```python
gender_counts = df['Gender'].value_counts()
colors = ['blue'] * len(gender_counts)
plt.pie(gender_counts,labels=['MALES','FEMALES'], colors=colors, autopct='%1.1f%%', startangle=140,wedgeprops={'edgecolor':'black'})
plt.title('Gender Distribution')
# we can see the the distribution of both genders is almost equal 
```

```python
df['Gender']=df['Gender'].astype('category')
```

```python
df['Age'].unique()
```

```python
sns.histplot(df['Age'], bins=20, color='skyblue', kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
# most of the age distribution lies between 40s and 50s 
# older people in general are more 
```

```python
customer_type_count= df['Customer Type'].value_counts()
```

```python
colors = ['blue', 'red']
plt.bar(customer_type_count.index,customer_type_count,color=colors)
plt.xlabel('type')
plt.ylabel('count')
plt.title('CUSTOMER_TYPE')
# more customers used our site/business before only a few new ones 
```

```python
sns.countplot(data=df, y='Type of Travel', color='blue')
plt.xlabel('Count')
plt.ylabel('Type of Travel')
plt.title('Type of Travel Distribution')
# most people travel beacuse of work or business to attend than people do for holidays or family  visits 
```

```python
class_counts = df['Class'].value_counts()
explode=[0,0,0.2]
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'red', 'aqua'],wedgeprops={'edgecolor':'black'}
      , explode=explode)
plt.title('Class Distribution')
# a small precentage of people travel on economy plus most prefer busisness or normal economy class 
```

```python
df['Class']=df['Class'].astype('category')
```

```python

age_groups = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60, 70, float('inf')], labels=['20-29', '30-39', '40-49', '50-59', '60-69', '70+'])


sns.countplot(data=df, x='Class', hue=age_groups,palette='Reds')

plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution by Age Group')

# older people take business class more than people in their 20s
```

```python
df['Arrival Delay'].fillna('UNKNOWN',inplace=True)
df['Arrival Delay'].isna().sum()
```

```python
average_ratings = df.groupby('Class')['Departure and Arrival Time Convenience'].mean()
colors = ['blue'] * len(average_ratings)
plt.pie(average_ratings, labels=average_ratings.index, autopct='%1.1f%%', startangle=90,wedgeprops={'edgecolor':'black'},colors=colors)
plt.title('AVARAGE RATINGS FOR Departure and Arrival Time Convenience')
# each class got an avarage 3.5 stars when it comes to Departure and Arrival Time with business class being the lowest with 3.2
```

```python
sns.countplot(data=df, x='Class', hue='Departure and Arrival Time Convenience',palette='RdPu')
# economy class has better ratings in departure and arrival time than other classes 
```

```python
ease_of_booking_counts = df['Ease of Online Booking'].value_counts()
plt.bar(ease_of_booking_counts.index,ease_of_booking_counts)
plt.xlabel('Ease of Online Booking RatingS')
plt.ylabel('Count')
df['Ease of Online Booking'].mean()
# it seems like people face diffculites using our website to book their flights it with avarage of 2.7 stars only 
```

```python
average_ratings = df.groupby('Class')['Check-in Service'].mean()
colors = ['blue'] * len(average_ratings)
plt.pie(average_ratings, labels=average_ratings.index, autopct='%1.1f%%', startangle=90,wedgeprops={'edgecolor':'black'},colors=colors)
plt.title('AVARAGE RATINGS FOR CHECK_IN_SERVICE')
# on avarage all classes are close in ratings when it comes down to Check-in Service with ratings of 3.1 to 3.6 appx
```
